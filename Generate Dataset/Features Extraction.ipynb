### Different port

```python
def diff_port(url):
    p = '(?:http.*://)?(?P<host>[^:/ ]+).?(?P<port>[0-9]*).*'

    m = re.search(p,url)
    if m.group('port') :
        return 1
    else :
        return 0
diff_port('http://www.abc.com:123/test')
```

```python
def append_row(url):
    global cld_dataset
    
    new_row = new_row = {'url':url, 'labels':1, 'type':'train'}
#     print(new_row)
    cld_dataset = cld_dataset.append(new_row,ignore_index=True)
    
```

```python
def main():
    
    data = [url for url in phishtank['url']]
    
    print([val for val in map(append_row, data)])
    with concurrent.futures.ProcessPoolExecutor() as executor:
#         print([val for val in executor.map(append_row, data)])
        future_proc = {executor.submit(append_row, url): url for url in data}
        for future in concurrent.futures.as_completed(future_proc):
            print(future.result())
        
if __name__ == '__main__':
    main()
```
